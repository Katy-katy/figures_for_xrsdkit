## Plots of all regresson models: True vs Predicted values

```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
import yaml
import pprint

%matplotlib inline
from util import get_true_predicted_crossvalid

import warnings
warnings.filterwarnings("ignore")
```

```python
from xrsdkit.tools.profiler import profile_keys
from xrsdkit.tools.ymltools import read_local_dataset
```

```python
df, idx_df = read_local_dataset('../xrsdkit_modeling/flowreactor_pd_nanoparticles/dataset', downsampling_distance=1) 
```

### Regression models for the System Class "Diffuse"

```python
diffuse_res = get_true_predicted_crossvalid('diffuse', df)
```

```python
for k, v in diffuse_res.items():
    fig, ax = plt.subplots(1,1, figsize=(7,4))
    x = v['true_y']
    y = v['cross_val']
    plt.scatter(x, y, marker='^')
    z = v['predicted']
    plt.scatter(x, z, marker='o')
    ax.set_title(k)
    ax.legend(['cross_val', 'predicted'])
```

### Regression models for the System Class "Diffuse_diffuse"

```python
dif_dif_res = get_true_predicted_crossvalid('diffuse__diffuse', df)
for k, v in dif_dif_res.items():
    fig, ax = plt.subplots(1,1, figsize=(7,4))
    x = v['true_y']
    y = v['cross_val']
    plt.scatter(x, y, marker='^')
    z = v['predicted']
    plt.scatter(x, z, marker='o')
    ax.set_title(k)
    ax.legend(['cross_val', 'predicted'])
```
